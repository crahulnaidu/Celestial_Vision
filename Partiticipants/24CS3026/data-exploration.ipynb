```python
# ================================
# SpaceNet Dataset Exploration
# Name: YOUR FULL NAME
# Enrollment Number: YOUR_ENROLLMENT_NO
# ================================

import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Root dataset path
DATASET_PATH = "/kaggle/input/spacenet-an-optimally-distributed-astronomy-data"

# Get actual class directory (inside main folder)
root_folder = os.listdir(DATASET_PATH)[0]
CLASS_PATH = os.path.join(DATASET_PATH, root_folder)

# -------------------------------
# 1. Load Classes
# -------------------------------
classes = [cls for cls in os.listdir(CLASS_PATH) 
           if os.path.isdir(os.path.join(CLASS_PATH, cls))]

print("Classes:", classes)
print("Number of classes:", len(classes))

# -------------------------------
# 2. Count Images Per Class
# -------------------------------
class_counts = {}

for cls in classes:
    class_dir = os.path.join(CLASS_PATH, cls)
    images = [f for f in os.listdir(class_dir)
              if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    class_counts[cls] = len(images)

df = pd.DataFrame.from_dict(class_counts, orient='index', columns=['Image Count'])
print("\nClass Distribution:")
display(df)

# -------------------------------
# 3. Plot Class Distribution
# -------------------------------
plt.figure(figsize=(10, 5))
sns.barplot(x=df.index, y=df["Image Count"])
plt.xticks(rotation=45)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()

# ================================
# SpaceNet Dataset Exploration
# Name: YOUR FULL NAME
# Enrollment Number: YOUR_ENROLLMENT_NO
# ================================

import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Root dataset path
DATASET_PATH = "/kaggle/input/spacenet-an-optimally-distributed-astronomy-data"

# Get actual class directory (inside main folder)
root_folder = os.listdir(DATASET_PATH)[0]
CLASS_PATH = os.path.join(DATASET_PATH, root_folder)

# -------------------------------
# 1. Load Classes
# -------------------------------
classes = [cls for cls in os.listdir(CLASS_PATH) 
           if os.path.isdir(os.path.join(CLASS_PATH, cls))]

print("Classes:", classes)
print("Number of classes:", len(classes))

# -------------------------------
# 2. Count Images Per Class
# -------------------------------
class_counts = {}

for cls in classes:
    class_dir = os.path.join(CLASS_PATH, cls)
    images = [f for f in os.listdir(class_dir)
              if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    class_counts[cls] = len(images)

df = pd.DataFrame.from_dict(class_counts, orient='index', columns=['Image Count'])
print("\nClass Distribution:")
display(df)

# -------------------------------
# 3. Plot Class Distribution
# -------------------------------
plt.figure(figsize=(10, 5))
sns.barplot(x=df.index, y=df["Image Count"])
plt.xticks(rotation=45)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()

# -------------------------------
# 4. Visualize Random Samples
# -------------------------------
plt.figure(figsize=(12, 8))

for i, cls in enumerate(classes):
    img_dir = os.path.join(CLASS_PATH, cls)
    img_name = random.choice(os.listdir(img_dir))
    img_path = os.path.join(img_dir, img_name)

    img = Image.open(img_path)
    plt.subplot(3, 3, i + 1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")

plt.tight_layout()
plt.show()

# -------------------------------
# 5. Image Resolution Analysis (Safe)
# -------------------------------
image_sizes = []

for cls in classes:
    img_dir = os.path.join(CLASS_PATH, cls)
    
    for img_name in os.listdir(img_dir):
        try:
            img = Image.open(os.path.join(img_dir, img_name))
            image_sizes.append(img.size)
        except:
            continue

sizes_df = pd.DataFrame(image_sizes, columns=["Width", "Height"])

print("\nImage Resolution Statistics:")
display(sizes_df.describe())

# -------------------------------
# 6. Observations
# -------------------------------
print("""
Observations:
- Dataset contains multiple astronomical object classes.
- Minor class imbalance exists.
- Images have varying resolutions.
- Dataset is suitable for CNN-based models after resizing.
""")
# -------------------------------
# 5. Image Resolution Analysis
# -------------------------------
image_sizes = []

for cls in classes:
    img_dir = os.path.join(CLASS_PATH, cls)
    for img_name in os.listdir(img_dir)[:10]:
        img = Image.open(os.path.join(img_dir, img_name))
        image_sizes.append(img.size)

sizes_df = pd.DataFrame(image_sizes, columns=["Width", "Height"])

print("\nImage Resolution Statistics:")
display(sizes_df.describe())

# -------------------------------
# 6. Observations
# -------------------------------
print("""
Observations:
- Dataset contains multiple astronomical object classes.
- Minor class imbalance exists.
- Images have varying resolutions.
- Dataset is suitable for CNN-based models after resizing.
""")
```
